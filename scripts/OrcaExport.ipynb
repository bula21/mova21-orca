```python
import urllib.request as urllib2
import pandas as pd
import glob
import os

ACTIVTY_URL_CSV = "https://orca.mova.ch/activity_executions.csv?activity_execution_filter%5Bdate%5D=&activity_execution_filter%5Bends_at_before%5D=&activity_execution_filter%5Bfield_id%5D=&activity_execution_filter%5Blanguage%5D=&activity_execution_filter%5Bmax_units%5D=&activity_execution_filter%5Bmin_available_headcount%5D=&activity_execution_filter%5Bspot_id%5D=&activity_execution_filter%5Bstarts_at_after%5D=&locale=de&activity_execution_filter%5Bactivity_id%5D="
ACTIVITIES = [ 122, 126, 187, 188, 243, 128, 118, 129, 130, 184, 124, 125, 123, 127, 185, 186, 83, 119, 120 ]
exportPath = "/content/"
headers = {'Cookie': '_orca_session=' }
```

```python
for activity in ACTIVITIES:
  print(ACTIVTY_URL_CSV + str(activity))

  request = urllib2.Request(ACTIVTY_URL_CSV + str(activity), headers=headers)
  request = urllib2.urlopen(request)
  with open(exportPath + 'orca_activity_export_' + str(activity) + '.csv','wb') as output:
    output.write(request.read())
```

```python
files = os.path.join(exportPath, "*.csv")
files = glob.glob(files)
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
df.to_csv(exportPath + 'orca_activity_export_' + "_".join([str(i) for i in ACTIVITIES ]) + ".csv")
```
